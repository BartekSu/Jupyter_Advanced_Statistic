```python
df = pd.read_csv('ADANIPORTS.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index(['Date'], inplace=True)
```

```python
wdf = df.resample('Y').mean()
#W -weekly
#M- Mounthly
#Y - Yearly
```
