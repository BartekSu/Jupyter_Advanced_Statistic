```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import f
x = np.linspace(0, 4.5, 1000)
data_l = f(1, 1, 0)
data_2 = f(5,8, 0)
data_3 = f(40, 40, 0)
data_4 = f(100, 200, 0)
plt.figure(figsize=(7, 5),dpi = 100) 
plt.plot(x,data_l.pdf(x), label = '1,1') 
plt.plot(x, data_2.pdf(x), label = '5,8') 
plt.plot(x, data_3.pdf(x), label = '40,40') 
plt.plot(x, data_4.pdf(x), label = '100,200') 
plt.legend()
```

 Null Hypotesis = Variance are qural 

 Alternative Hypotesis are not equal

```python
import random
random.seed(20)
```

```python
x = np.array([random.gauss(100,15) for x in range(20)])
y = np.array([random.gauss(100,15) for x in range(20)])
```

```python
f_test_stat = np.var(x,ddof=1) / np.var(y,ddof = 1)
f_test_stat
```

```python
dfn = x.size - 1
dfd = y.size - 1
print(dfn)
print(dfd)
```

```python
p_value = 1 - f.cdf(f_test_stat,dfn,dfd)
p_value
```

```python
x = np.array([random.gauss(100,15) for x in range(20)])
y = np.array([random.gauss(200,15) for x in range(20)])
```

```python
f_test_stat = np.var(x,ddof=1) / np.var(y,ddof = 1)
print('f_test_stat=',f_test_stat)
dfn = x.size - 1
dfd = y.size - 1
print('dfn=',dfn)
print('dfd=',dfd)
p_value = 1 - f.cdf(f_test_stat,dfn,dfd)
print('p_value=',p_value)
```
